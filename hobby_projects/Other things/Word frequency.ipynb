```python
import requests
import re
from collections import Counter
from nltk.corpus import stopwords
import nltk

# Make sure to run once:
nltk.download('stopwords')
```

```python
def get_random_article_text():
    summary = requests.get("https://en.wikipedia.org/api/rest_v1/page/random/summary").json()
    title = summary['title']

    params = {
        "action": "query",
        "prop": "extracts",
        "format": "json",
        "explaintext": True,
        "titles": title
    }

    response = requests.get("https://en.wikipedia.org/w/api.php", params=params)
    pages = response.json()["query"]["pages"]
    page = next(iter(pages.values()))
    return title, page.get("extract", "")

def count_word_frequencies(text):
    words = re.findall(r'\b\w+\b', text.lower())
    stop_words = set(stopwords.words("english"))
    filtered = [word for word in words if word not in stop_words]
    return Counter(filtered)

title, text = get_random_article_text()
freqs = count_word_frequencies(text)

print(f"\nTop words in article: {title}\n")
for word, count in freqs.most_common(10):
    print(f"{word}: {count}")
```
